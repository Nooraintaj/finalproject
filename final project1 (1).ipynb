# TOKENIZING

```python
clean_text_2=[]
from nltk.tokenize import sent_tokenize, word_tokenize

import nltk
nltk.download('punkt')
```

```python
sent_tok=[]
for sent in clean_text_1:
    sent=sent_tokenize(sent)
    sent_tok.append(sent)
sent_tok
```

# WORD TOKENIZING 

```python
clean_text_2=[word_tokenize(i) for i in clean_text_1]
clean_text_2
```

```python
import re
clean_text_3=[]

for word in clean_text_2:
    clean=[]
    for w in word:
        res=re.sub(r'[^\w\s]','',w)
        if res != '':
            clean.append(res)
    clean_text_3.append(clean)
clean_text_3            
```

# STOP WORD REMOVAL

```python
import nltk
nltk.download("stopwords")
```

```python
from nltk.corpus import stopwords
```

```python
clean_text_4=[]
for words in clean_text_3:
    w=[]
    for word in words:
        if not word in stopwords.words('english'):
            w.append(word)
    clean_text_4.append(w)         
```

```python
clean_text_4
```

# STEMMING

```python
#stemming converts the data into its root form
from nltk.stem.porter import PorterStemmer
port= PorterStemmer()
```

```python
clean_text_5=[]
for words in clean_text_4:
    w=[]
    for word in words:
        w.append(word)
    clean_text_5.append(w)
clean_text_5    
```

# LIMITIZING

```python
from nltk.stem.wordnet import WordNetLemmatizer
wnet=WordNetLemmatizer()
```

```python
import nltk 
nltk.download('wordnet')
```

```python
lem=[]
for words in clean_text_4:
    w=[]
    for word in words:
        w.append(wnet.lemmatize(word))
    lem.append(w)
lem
```
